### 2. ModÃ¨le Ã  base de rÃ¨gles

Utilisez les box-plots des variables qui ont une majoritÃ© de valeurs diffÃ©rentes pour
les diffÃ©rentes classes afin de programmer des rÃ¨gles (if-then-else) permettant la

```python
import seaborn as sns

def plot_carac_as_boxplot(carac, source):
    
    firstClass = source.loc[[1]]
    secondClass = source.loc[[2]]
    thirdClass = source.loc[[3]]
    return [firstClass[carac], secondClass[carac], thirdClass[carac]]
```

Nous commenÃ§ons tout d'abord par analyser la rÃ©partition des donnÃ©es selon les catÃ©gories. Nous ciblons pour ce faire chacune
des caractÃ©ristiques Ã  l'aide de boxplots.

Ceci nous permettra par la suite de sÃ©lectionner des caractÃ©ristiques dont les valeurs diffÃ¨rent fortement selon les classes.
Nous pourrons ainsi Ã©tablir nos rÃ¨gles en nous basant sur ces derniÃ¨res.

```python
caracs = raw_data.keys()
import math
lines = 4
cols = 4

fig, ax = plt.subplots(lines, cols, figsize=(18,16))
for x in range(len(caracs)):
    # print("x = %d (%s), line = %f, col = %f" % (x,caracs[x], math.floor(x / 4), (x % 4)))
    ax[math.floor(x / 4), x % 4].set_title(caracs[x])
    ax[math.floor(x / 4), x % 4].boxplot(plot_carac_as_boxplot(caracs[x], raw_data))
```

sur la base des boxplots ci-dessus nous pouvons tirer les conclusions suivantes :

* les caractÃ©ristiques "Flavanoids", "Total phenols", "Alcohol" et eventuellement "Nonflavanoid phenols" sont de bons candidats sur lequels appuyer les rÃ¨gles de notre modÃ¨le.

En effet nous pouvons remarquer que de maniÃ¨re gÃ©nÃ©rale les plages de valeurs observÃ©es pour ces caractÃ©ristiques semblent Ãªtre distinctes selon les classes et ne se chevauchent que peu.

* Au contraire, les caractÃ©ristiques "Magnesium" et "Ash" seront difficilement exploitables par notre modÃ¨le

Ces deux caractÃ©ristiques ont des plages de valeurs qui se chevauchent fortement voir se confondent entre les classes. Il sera donc difficile d'identifier de maniÃ¨re correcte la classe d'une donnÃ©e en se basant sur ces valeurs.

Nous allons donc tout d'abord crÃ©er 3 modÃ¨les dotÃ© d'une seule rÃ¨gle permettant de discriminer les donnÃ©es sur la base d'une caractÃ©ristiques Ã  la fois :

* "Alcohol"
* "Flavanoids"

```python
from sklearn.metrics import confusion_matrix
# See: https://matplotlib.org/stable/gallery/images_contours_and_fields/image_annotated_heatmap.html

classes = raw_data.index

def plot_confusion_matrix(predictions_results):
    
    conf_matrix = confusion_matrix(classes, predictions_results)
    
    fig, ax = plt.subplots()
    im = ax.imshow(conf_matrix, cmap="viridis")

    # Show all ticks and label them with the respective list entries
    ax.set_xticks(np.arange(len(classes_names)), labels=classes_names)
    ax.set_yticks(np.arange(len(classes_names)), labels=classes_names)

    # Rotate the tick labels and set their alignment.
    plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")

    # Loop over data dimensions and create text annotations.
    for i in range(len(classes_names)):
        for j in range(len(classes_names)):
            text = ax.text(j, i, conf_matrix[i, j], fontweight="bold", fontsize="large", ha="center", va="center", c="r")

    fig.tight_layout()
    plt.show()
    
def predict_class(compare, colName):
    actual = raw_data[colName].to_numpy()
    predictions = np.full(len(actual), 0)
    
    for i in range(len(actual)):
        predictions[i] = compare(actual[i])
    return predictions
```

```python
# ModÃ¨le + rÃ¨gle basÃ©e sur la variable "Flavanoids"

def compare(i):
    if(i < 1.5):
        return 3
    else:
        return 1 if i > 2.5 else 2

predictions_results = predict_class(compare, "Flavanoids")
plot_confusion_matrix(predictions_results)
```

```python
# ModÃ¨le + rÃ¨gle basÃ©e sur la variable "Alcohol"

def compare(i):
    if(i > 13.5):
        return 1
    else:
        return 2 if i < 12.5 else 3

predictions_results = predict_class(compare, "Alcohol")
plot_confusion_matrix(predictions_results)
```

### 5. Lâ€™algorithme LVQ

#### # mÃ©thodes fournies
rappels :
* codebook = ensemble des vecteurs prototypes
* 2 phases pour le crÃ©er : init des vecteurs prot. / mise Ã  jour progressive des vecteurs prot. sur n epochs

```python
# ---------------------- codebook management

# Create a random codebook vector
def init_codevector(train,category):
    n_records = len(train)
    n_features = len(train[0])-1
    found = False
    while(not found):
        random_observation = randrange(n_records)
        if (train[random_observation][-1] == category):
            found = True

    codevector = [train[random_observation][i] for i in range(n_features)]
    codevector.append(category)
    return codevector

# Create a random codebook vector
def random_codebook(train):
    n_records = len(train)
    n_features = len(train[0])
    codebook = [train[randrange(n_records)][i] for i in range(n_features)]
    return codebook

# Train a set of codebook vectors
def train_codebook(train, n_codevectors, lrate, epochs):
    codebook = [random_codebook(train) for i in range(n_codevectors)]
    for epoch in range(epochs):
        rate = lrate * (1.0-(epoch/float(epochs)))
        sum_error = 0.0
        shuffle(train)
        for row in train:
            # bmu => prototype le + proche du point saisi alÃ©atoirement
            bmu = get_best_matching_unit(codebook, row)
            for i in range(len(row)-1):
                error = row[i] - bmu[i]
                sum_error += error**2
                # s'ils sont de la mÃªme classe (derniÃ¨re colonne catÃ©gorie semblable) on rapproche le prototype
                if bmu[-1] == row[-1]:
                    bmu[i] += rate * error
                else:
                    # sinon on l'Ã©loigne
                    bmu[i] -= rate * error
        print('>epoch=%d, lrate=%.3f, error=%.3f' % (epoch, rate, sum_error))
    return codebook


# Calculate the Euclidean distance between two vectors
def euclidean_distance(row1, row2):
    distance = 0.0
    for i in range(len(row1)-1):
        distance += (row1[i] - row2[i])**2
    return sqrt(distance)

# Locate the best matching unit
def get_best_matching_unit(codebook, test_row):
    distances = list()
    for codevector in codebook:
        dist = euclidean_distance(codevector, test_row)
        distances.append((codevector, dist))
    distances.sort(key=lambda tup: tup[1])
    return distances[0][0]

# Make a prediction with codebook vectors
def predict(codebooks, test_row):
    bmu = get_best_matching_unit(codebooks, test_row)
    return bmu[-1]

def predict_lvq(train, test, n_codebooks, lrate, epochs):
    codebooks = train_codebook(train, n_codebooks, lrate, epochs)
    return [predict(codebooks, row) for row in test]
```
